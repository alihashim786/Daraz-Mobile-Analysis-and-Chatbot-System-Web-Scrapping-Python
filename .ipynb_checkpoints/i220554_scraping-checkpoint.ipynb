```python
#Importing the necessary libraries

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = 'https://www.daraz.pk/'
driver.get(url)

# Find the main grid containing categories
elec = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div[2]/div/div/div/div/div/div/div/ul/li[7]')
elec.click()
smart = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div[2]/div/div/div/div/div/div/div/ul/ul[7]/li[1]')
smart.click()  
#next_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[9]/a')
```

```python
import requests
import time
import pandas as pd
import numpy as np
import re

Current_Daraz_Page = 1
data = []
specifications = []
reviews_data = []
UniqueID = 0

# Define the base URL of the search results page
base_url = "https://www.daraz.pk/smartphones/?spm=a2a0e.home.cate_7.1.35e34076JHwMAq"


# Loop over first 5 pages
for i in range(1, 6):
    page_url = f"{base_url}&page={page}"
    driver.get(page_url)
    top = 0
    bottom = 315
    #find the ad grid and ad cards on the page  
    ad_grid = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[2]')
    ad_cards = ad_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')
    for i, mobile in enumerate(ad_cards):
        UniqueID = UniqueID + 1
        try:
            title_element = mobile.find_element(By.CSS_SELECTOR, '#id-a-link')
            url = title_element.get_attribute('href')
            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[1])
            driver.get(url)
            time.sleep(1)
            #Description and Brand
            try:
                description = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[3]/div/div/span').text
            except NoSuchElementException:
                description = "Not Found"
            
            brand = description.split(' ')[0]
            
            #Price
            try:
                price = driver.find_element(By.XPATH,'/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[7]/div/div/span').text
                price = price[4:]
                price = price.replace(',', '')
            except NoSuchElementException:
                price = "0"


            time.sleep(1)
            driver.execute_script("window.scrollTo(0, 500)")
            time.sleep(1) 
            
                       
            #Get rating out of 5
            try:
                Rating_element = driver.find_element(By.CSS_SELECTOR, '.review-info-rate .score')
                Rating = Rating_element.text
            except NoSuchElementException:
                Rating = 0.0
           
            driver.execute_script(f"window.scrollTo(500, 1000)")
            
            ReviewID = 0
            #GETTING REVIEWS OF THE FIRST 2 PAGES
            for page in range(0, 2):
                review = []

                try:
                    reviews_grid = driver.find_element(By.CSS_SELECTOR, '#module_product_review')
                except NoSuchElementException:
                    print('Reviews Not Avaialable')
                    break

                reviews = reviews_grid.find_elements(By.CLASS_NAME, 'review-item')

                for i in range(len(reviews)):
                    if reviews[i].find_element(By.CSS_SELECTOR, '.review-content-sl').text != '':
                        ReviewContent = reviews[i].find_element(By.CSS_SELECTOR, '.review-content-sl').text
                        ReviewName = reviews[i].find_element(By.CLASS_NAME, 'user').text
                        ReviewID = ReviewID + 1

                        #Now counting the stars of ratings
                        Stars = reviews[i].find_elements(By.XPATH, './/img[contains(@src, "//laz-img-cdn.alicdn.com/tfs/TB19ZvEgfDH8KJjy1XcXXcpdXXa-64-64.png")]')
                        RatingOfReview = len(Stars)

                        review.append({'UID': UniqueID, 'Review_ID': ReviewID, 'Name': ReviewName, 'Rating': RatingOfReview, 'Text': ReviewContent})

                reviews_data.append(review)
                
            #Check Specs
            driver.execute_script("window.scrollTo(1000, 2000)")
            top2 = 1000    
            specs_dict = {'UID': UniqueID}
            bottom2 = 1500
            
            try:
                while True:
                    specs_grid = driver.find_elements(By.CSS_SELECTOR, '.key-li')
                    MoreButton = None
                    try:
                        driver.execute_script(f"window.scrollTo({top2}, {bottom2})")
                        MoreButton = driver.find_element(By.CSS_SELECTOR, '.expand-button button.pdp-view-more-btn')
                        MoreButton.click()
                        top2 = bottom2
                        time.sleep(1)
                        bottom2 = bottom2 + 500
                    except:
                        pass

                    if not specs_grid and not MoreButton:
                        break

                    for spec in specs_grid:
                        Spec_Detail = spec.find_element(By.CSS_SELECTOR, '.html-content.key-value').text
                        SpecTitle = spec.find_element(By.CLASS_NAME, 'key-title').text
                        specs_dict[SpecTitle] = Spec_Detail  # Append to the existing dictionary

                    if specs_grid:
                        break

            except (NoSuchElementException, StaleElementReferenceException):
                print('No specs')

            # Store specs in a list
            specifications.append(specs_dict)
            
            
            
            try:
                driver.execute_script(f"window.scrollTo({top2}, {bottom})")
                QuestionText = driver.find_element(By.XPATH, '//*[@id="module_product_qna"]/div/div[1]/div').text
                Matched = re.search(r'\((\d+)\)', QuestionText)
                
                # Extracting the number from matched group
                if Matched:
                    QuestionNumber = Matched.group(1)
            except (NoSuchElementException, StaleElementReferenceException):
                print('ERROR FOUND')
                
                
                
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            #append to data list
            data.append({'UID': UID, 'Name': description, 'Brand': brand, 'Price': price, 'Rating': Rating, 'Questions Count' : QuestionNumber ,'URL' : url})
        
        
        except Exception as e:
            print(f"ERROR IN PROCESSING MOBILE ADD CARD: {e}")
        
        
        #Ta k 4 mobiles k baad scroll karay
        if i % 3 == 0:
            driver.execute_script(f"window.scrollTo({top},{bottom})")
            top = bottom
            bottom = bottom + 315
            time.sleep(1)

      
    while Current_Daraz_Page < 6:
    # Your code here
        Current_Daraz_Page = Current_Daraz_Page + 1  # Increment daraz_page in your code logic
    
    next_button =   driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[9]')  
    next_button.click()
    time.sleep(1)
```

```python
df1 = pd.DataFrame(data)
df1.to_csv('Essential_Info.csv')
df2 = pd.DataFrame(reviews_data)
df2.to_csv('Reviews.csv')
df3 = pd.DataFrame(specs_data)
df1.to_csv('Specifications.csv')
AllData = pd.concat([df1,df2,df3], axis = 1)
df = pd.DataFrame(AllData)
df.to_csv("ALL_DATA.csv")
```
